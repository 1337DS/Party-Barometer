```python
dict_termine={}
list=soup.find_all('div',{'class': 'tx-uma-portal-exam-category'})
for i in list:
  print(i.find('h2').text)
  dates=set()
  # hier nehmen wir set (Menge) weil manche termine doppelt sind
  for t in (i.find_all('td')):
      if "2022" in t.text:
        print(t.text)
        dates.add(t.text)
  dict_termine[i.find('h2').text]=dates
  
```
