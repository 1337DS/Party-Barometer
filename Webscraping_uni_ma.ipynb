```python
from bs4 import BeautifulSoup
import requests
```

```python
URL = 'https://www.uni-mannheim.de/pruefungstermine/pruefungen-ersttermine-bachelor/' 
website = requests.get(URL)
```

```python
soup = BeautifulSoup(website.text, 'html.parser')
```

```python
soup.find_next_sibling('option')
```

```python
soup.find_all('option')
```

```python
soup.find_all('td')
```

```python
website.text
```

```python
for i in soup.find_all('data-category'):
  print(i.findChildren)
```

```python
li = soup.find('div',{'class': 'tx-uma-portal-exam-category'})
children = li.findChildren("td")
for child in children:
    print(child)
```

```python
list=soup.find_all('div',{'class': 'tx-uma-portal-exam-category'})
for i in list:
  print(i.find('h2'))
  for t in (i.find_all('td')):
    if "2022" in t.text:
      print(t.text)
```
